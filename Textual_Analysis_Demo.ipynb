# Textual Analysis Demo

Comprehensive demonstration of sentiment analysis using **Loughran McDonald**, **VADER**, **Opinion Lexicon**, and **Readability Metrics** (FOG & Flesch-Kincaid)

## What you'll learn:
1. How to initialize the TextualAnalyzer
2. Analyze individual texts
3. Process entire Excel columns
4. Interpret sentiment scores
5. Compare readability metrics
6. Visualize results

## Section 1: Import Libraries and Initialize

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textual_analysis import TextualAnalyzer

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("✓ Libraries imported successfully")
```

```python
# Initialize the analyzer
analyzer = TextualAnalyzer()
print("✓ TextualAnalyzer initialized")
print("✓ Loaded: Loughran McDonald, VADER, Opinion Lexicon")
print("✓ Ready for analysis")
```

## Section 2: Create Sample Data

```python
# Sample texts with different sentiments and complexities
sample_texts = [
    "The company reported strong profit growth this quarter.",
    "Sales declined significantly due to market uncertainty and competition.",
    "We expect positive earnings momentum in the coming months.",
    "The implementation of comprehensive strategic initiatives necessitates meticulous analysis.",
    "Customer satisfaction is excellent and sales are good.",
    "Risk factors include regulatory changes, litigation, and market volatility.",
    "Amazing product! I absolutely love it!!!",
    "Terrible experience. Very disappointed with the service."
]

# Create a DataFrame
df_sample = pd.DataFrame({
    'text_id': range(1, len(sample_texts) + 1),
    'text': sample_texts
})

print("Sample Data Created:")
print(df_sample)

# Save to Excel for demonstration
df_sample.to_excel('sample_data.xlsx', index=False)
print("\n✓ Sample data saved to 'sample_data.xlsx'")
```

## Section 3: Analyze Individual Texts

```python
# Analyze first text in detail
text1 = sample_texts[0]
print(f"Text: {text1}")
print("\n" + "="*80)

results1 = analyzer.analyze_text(text1)

# Organize results by category
print("\n📊 LOUGHRAN MCDONALD (Financial Sentiment):")
print("-" * 80)
for key, value in results1.items():
    if key.startswith('LM_'):
        print(f"  {key:40s}: {value:8.4f}")

print("\n💬 VADER (Social Media Sentiment):")
print("-" * 80)
for key, value in results1.items():
    if key.startswith('VADER_'):
        print(f"  {key:40s}: {value:8.4f}")

print("\n📚 OPINION LEXICON (Harvard-based):")
print("-" * 80)
for key, value in results1.items():
    if key.startswith('Opinion_'):
        print(f"  {key:40s}: {value:8.4f}")

print("\n📖 READABILITY METRICS:")
print("-" * 80)
for key, value in results1.items():
    if key.startswith(('FOG_', 'Flesch_')):
        print(f"  {key:40s}: {value:8.4f}")
```

## Section 4: Analyze All Sample Texts

```python
# Analyze all texts
results_list = []
for idx, text in enumerate(sample_texts):
    print(f"Processing text {idx + 1}/{len(sample_texts)}", end='\r')
    results = analyzer.analyze_text(text)
    results_list.append(results)

# Create results DataFrame
df_results = pd.DataFrame(results_list)
df_combined = pd.concat([df_sample, df_results], axis=1)

print(f"\n✓ Analysis complete! Total columns: {len(df_combined.columns)}")

# Display summary
print("\nColumns in output:")
for i, col in enumerate(df_combined.columns, 1):
    print(f"{i:2d}. {col}")
```

## Section 5: Sentiment Score Comparison

```python
# Compare sentiment scores across dictionaries
sentiment_comparison = df_combined[[
    'text',
    'LM_net_sentiment',
    'VADER_compound',
    'Opinion_Lexicon_sentiment_score'
]].copy()

sentiment_comparison.columns = [
    'Text',
    'Loughran McDonald',
    'VADER',
    'Opinion Lexicon'
]

print("\n📊 SENTIMENT SCORE COMPARISON:")
print("=" * 120)
print(sentiment_comparison.to_string(index=False))

# Save comparison
sentiment_comparison.to_csv('sentiment_comparison.csv', index=False)
print("\n✓ Saved to 'sentiment_comparison.csv'")
```

## Section 6: Readability Metrics Analysis

```python
# Readability comparison
readability_comparison = df_combined[[
    'text',
    'FOG_index',
    'FOG_complexity',
    'Flesch_Kincaid_Grade'
]].copy()

print("\n📖 READABILITY METRICS:")
print("=" * 120)
print(readability_comparison.to_string(index=False))

# Interpretation
print("\n📚 INTERPRETATION GUIDE:")
print("-" * 120)
print("FOG Index & Flesch-Kincaid interpretation:")
print("  6-8 :  Easy to read (elementary/middle school level)")
print("  9-12:  Medium difficulty (high school level)")
print("  13+ :  Difficult (college/graduate level)")
```

## Section 7: Visualizations

```python
# Visualization 1: Sentiment Scores Comparison
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Loughran McDonald
axes[0].barh(range(len(df_combined)), df_combined['LM_net_sentiment'], color='steelblue')
axes[0].set_yticks(range(len(df_combined)))
axes[0].set_yticklabels([f"Text {i+1}" for i in range(len(df_combined))])
axes[0].set_xlabel('Net Sentiment (Positive - Negative)')
axes[0].set_title('Loughran McDonald\n(Financial Sentiment)', fontweight='bold')
axes[0].axvline(x=0, color='red', linestyle='--', alpha=0.5)

# VADER
colors = ['red' if x < -0.05 else 'gray' if -0.05 <= x <= 0.05 else 'green' 
          for x in df_combined['VADER_compound']]
axes[1].barh(range(len(df_combined)), df_combined['VADER_compound'], color=colors)
axes[1].set_yticks(range(len(df_combined)))
axes[1].set_yticklabels([f"Text {i+1}" for i in range(len(df_combined))])
axes[1].set_xlabel('Compound Score (-1 to +1)')
axes[1].set_title('VADER\n(Social Media Sentiment)', fontweight='bold')
axes[1].axvline(x=0, color='red', linestyle='--', alpha=0.5)
axes[1].set_xlim(-1.1, 1.1)

# Opinion Lexicon
colors = ['red' if x < -0.05 else 'gray' if -0.05 <= x <= 0.05 else 'green' 
          for x in df_combined['Opinion_Lexicon_sentiment_score']]
axes[2].barh(range(len(df_combined)), df_combined['Opinion_Lexicon_sentiment_score'], color=colors)
axes[2].set_yticks(range(len(df_combined)))
axes[2].set_yticklabels([f"Text {i+1}" for i in range(len(df_combined))])
axes[2].set_xlabel('Sentiment Score (-1 to +1)')
axes[2].set_title('Opinion Lexicon\n(Harvard IV-4 Based)', fontweight='bold')
axes[2].axvline(x=0, color='red', linestyle='--', alpha=0.5)
axes[2].set_xlim(-1.1, 1.1)

plt.tight_layout()
plt.savefig('sentiment_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Sentiment comparison chart saved")
```

```python
# Visualization 2: Readability Metrics
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# FOG Index
axes[0].barh(range(len(df_combined)), df_combined['FOG_index'], color='coral')
axes[0].set_yticks(range(len(df_combined)))
axes[0].set_yticklabels([f"Text {i+1}" for i in range(len(df_combined))])
axes[0].set_xlabel('FOG Index (Years of Education)')
axes[0].set_title('FOG Index\n(Gunning FOG - Readability)', fontweight='bold')
axes[0].axvline(x=9, color='green', linestyle='--', alpha=0.5, label='High School (9)')
axes[0].axvline(x=13, color='orange', linestyle='--', alpha=0.5, label='College (13)')
axes[0].legend()

# Flesch-Kincaid
axes[1].barh(range(len(df_combined)), df_combined['Flesch_Kincaid_Grade'], color='skyblue')
axes[1].set_yticks(range(len(df_combined)))
axes[1].set_yticklabels([f"Text {i+1}" for i in range(len(df_combined))])
axes[1].set_xlabel('Grade Level')
axes[1].set_title('Flesch-Kincaid Grade Level\n(US School Grade)', fontweight='bold')
axes[1].axvline(x=9, color='green', linestyle='--', alpha=0.5, label='High School (9)')
axes[1].axvline(x=13, color='orange', linestyle='--', alpha=0.5, label='College (13)')
axes[1].legend()

plt.tight_layout()
plt.savefig('readability_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Readability comparison chart saved")
```

```python
# Visualization 3: Word Count Categories
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# LM Categories
lm_categories = df_combined[[
    'LM_positive_count', 'LM_negative_count', 'LM_uncertainty_count', 'LM_litigious_count'
]]
lm_categories.columns = ['Positive', 'Negative', 'Uncertainty', 'Litigious']
lm_categories.plot(kind='bar', ax=axes[0, 0], color=['green', 'red', 'orange', 'purple'])
axes[0, 0].set_title('Loughran McDonald Word Counts', fontweight='bold')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_xlabel('Text')
axes[0, 0].legend(loc='upper right')

# VADER Proportions
vader_props = df_combined[['VADER_negative', 'VADER_neutral', 'VADER_positive']]
vader_props.columns = ['Negative', 'Neutral', 'Positive']
vader_props.plot(kind='bar', ax=axes[0, 1], color=['red', 'gray', 'green'], stacked=False)
axes[0, 1].set_title('VADER Sentiment Proportions', fontweight='bold')
axes[0, 1].set_ylabel('Proportion')
axes[0, 1].set_xlabel('Text')
axes[0, 1].legend(loc='upper right')

# Opinion Lexicon Counts
ol_counts = df_combined[['Opinion_Lexicon_positive_count', 'Opinion_Lexicon_negative_count']]
ol_counts.columns = ['Positive', 'Negative']
ol_counts.plot(kind='bar', ax=axes[1, 0], color=['green', 'red'])
axes[1, 0].set_title('Opinion Lexicon Word Counts', fontweight='bold')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_xlabel('Text')
axes[1, 0].legend(loc='upper right')

# FOG Complexity
axes[1, 1].scatter(df_combined['FOG_index'], df_combined['FOG_complexity'], s=200, alpha=0.6)
axes[1, 1].set_xlabel('FOG Index')
axes[1, 1].set_ylabel('FOG Complexity (%)')
axes[1, 1].set_title('FOG Index vs Complexity', fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('detailed_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Detailed analysis chart saved")
```

## Section 8: Export Complete Analysis

```python
# Save complete results to Excel
df_combined.to_excel('complete_analysis_results.xlsx', index=False)
print("✓ Complete analysis saved to 'complete_analysis_results.xlsx'")

# Summary statistics
print("\n" + "="*80)
print("SUMMARY STATISTICS")
print("="*80)

summary_cols = [
    'LM_net_sentiment', 'VADER_compound', 'Opinion_Lexicon_sentiment_score',
    'FOG_index', 'FOG_complexity', 'Flesch_Kincaid_Grade'
]

summary = df_combined[summary_cols].describe().round(4)
print(summary)
```
